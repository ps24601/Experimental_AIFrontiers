```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
sns.set_style("darkgrid")
```

```python
deltaT = 10000
start = 100
```

```python
def brownian(start, mu, sig):
    n = 10000
    dt = 1/deltaT
    steps = [start]
    for i in range(n):
        deltaW = np.random.normal(0,np.sqrt(dt))
        nextS = steps[-1] + mu*steps[-1]*dt + sig*deltaW*steps[-1]
        steps.append(nextS)

    return steps
```

```python
arrayB = brownian(100,0.2,0.5)
plt.figure(figsize= (20,6))
plt.plot(arrayB, markersize = 0.1)
```

```python
mu = 0.2
sig = 0.5
arrayB = []
for i in range(500):
    arrayB.append(brownian(100,mu=mu, sig=sig))
arrayB = np.array(arrayB)
    
```

```python
indexT = int(0.8*(arrayB.shape[1]))
timewindowArray = arrayB[:,indexT]
np.mean(timewindowArray > 120)
```
